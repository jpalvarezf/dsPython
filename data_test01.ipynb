```python
import pandas as pd
import seaborn as sns
import numpy as np
```

#### Obtenemos los datos

```python
allstate = pd.read_csv("../data/allstate.csv")
```

* Manejo de datos

```python
cuant=pd.Series(['day',
                 'group_size',
                 'car_age',
                 'age_oldest',
                 'age_youngest',
                 'cost'])
stats=allstate[cuant].describe()
estadisticos=stats.index.values

for i in range(len(cuant)):
    y=estadisticos
    z=[cuant[i],cuant[i],cuant[i],cuant[i],cuant[i],cuant[i],cuant[i],cuant[i]]
    x=stats[cuant[i]]
    new=pd.DataFrame({"var_name": z,
                        "estadistico": y, 
                        "value": x})
    if i==0:
        data=new
    else:
        data=data.append(new)

data=data.set_index(np.arange(0,len(data)))
data=data[["var_name","estadistico","value"]]
```
